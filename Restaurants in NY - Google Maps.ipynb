# Import required libraries

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from time import sleep
```

# Open the browser and Navigate the webpage

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
driver.get('https://www.google.com/search?q=restaurants+in+new+york+usa+google+maps&sca_esv=601425548&rlz=1C1CHBD_enIN1020IN1020&biw=1366&bih=633&tbm=lcl&sxsrf=ACQVn0-FrJxG3ZjDv7qFavNr1LC1_sgUhw%3A1706201915367&ei=O5OyZbqJFc_V4-EP3NWWyAk&oq=restaurants+in+new+york+usa+google+&gs_lp=Eg1nd3Mtd2l6LWxvY2FsIiNyZXN0YXVyYW50cyBpbiBuZXcgeW9yayB1c2EgZ29vZ2xlICoCCAAyBRAhGKABMgUQIRigAUiJOFDzEVi2KHAAeACQAQCYAYUCoAG-DqoBBTAuNC41uAEByAEA-AEBwgIEECMYJ8ICBhAAGBYYHsICCxAAGIAEGIoFGIYDwgIHECEYChigAYgGAQ&sclient=gws-wiz-local#rlfi=hd:;si:;mv:[[40.7717958,-73.9745172],[40.705006499999996,-74.0162524]];tbs:lrf:!1m4!1u3!2m2!3m1!1e1!2m1!1e3!3sIAE,lf:1,lf_ui:4')
sleep(5)
driver.maximize_window()
```

```python
names = []
add = []
num = []
web = []
direct = []
```

```python
driver.refresh()
for i in range(4, 46, 2):
    try:
        rest = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[9]/div[1]/div/div[2]/div[2]/div/div/div/div/div/div/div/div/div[1]/div[3]/div[{0}]/div[2]/div/div/a'.format(str(i)))
        rest.click()
    except:
        print(" ")
            
    try:
        name = driver.find_element(By.XPATH,'//h2/span')
        names.append(name.text)
    except:
        names.append(" ")
            
    try: 
        address = driver.find_element(By.XPATH,'//span[@class="LrzXr"]')
        add.append(address.text)
    except:
        add.append(" ")
            
    try:
        number = driver.find_element(By.XPATH,'//span[@class="LrzXr zdqRlf kno-fv"]/a/span')
        num.append(number.text)
    except:
        num.append(" ")
        
    try: 
        website = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[9]/div[1]/div/div[2]/div[2]/div/div/div/div/div/div/div/div/div[1]/div[3]/div[{0}]/div[2]/div/a[1]'.format(str(i))).get_attribute('href')
        web.append(website)
    except:
        web.append(" ")
            
    try:
        direction = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[9]/div[1]/div/div[2]/div[2]/div/div/div/div/div/div/div/div/div[1]/div[3]/div[{0}]/div[2]/div/a[2]'.format(str(i))).get_attribute('href')
        direct.append(direction)
    except:
        direct.append(" ")
        
    sleep(3)   
```

```python
data = pd.DataFrame()
data['Restaurants Name'] = names
data['Address'] = add
data['Mob. Number'] = num
data['Website url'] = web
data['Location link'] = direct
```

```python
data['Website url'] = data['Website url'].shift(1)
data['Location link'] = data['Location link'].shift(1)
```

```python
datas = data.drop(data.index[0])
```
